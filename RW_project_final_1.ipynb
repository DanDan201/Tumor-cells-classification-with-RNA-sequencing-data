```python
q = open('data_T.csv')
data = pd.read_csv(q)

data
```

```python
#https://dataanalyticsireland.ie/2021/05/03/how-to-change-the-headers-on-a-csv-file/
#https://thispointer.com/edit-cell-values-in-csv-files-using-pandas-in-python/

data.loc[0,'gene_ID'] = "control"
data.loc[1,'gene_ID'] = "control"
data.loc[2,'gene_ID'] = "control"
data.loc[3,'gene_ID'] = "control"
data.loc[4,'gene_ID'] = "control"
data.loc[5,'gene_ID'] = "GBM"
data.loc[6,'gene_ID'] = "GBM"
data.loc[7,'gene_ID'] = "GBM"
data.loc[8,'gene_ID'] = "GBM"
data.loc[9,'gene_ID'] = "GBM"
data.loc[10,'gene_ID'] = "meta"
data.loc[11,'gene_ID'] = "meta"
data.loc[12,'gene_ID'] = "meta"
data.loc[13,'gene_ID'] = "meta"
data.loc[14,'gene_ID'] = "meta"
data.loc[15,'gene_ID'] = "control"
data.loc[16,'gene_ID'] = "meta"

data.rename(columns={'gene_ID': 'Target'}, inplace=True)

data
```

```python
X = data.loc[:, data.columns != 'Target']
X
```

```python
Y = data.loc[:, data.columns == 'Target']
Y
```

```python
#https://stats.stackexchange.com/questions/402470/how-can-i-use-scaling-and-log-transforming-together
import numpy as np

X_log = np.log(1+X)
X_log
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.LeaveOneOut.html
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html
#https://numpy.org/doc/stable/reference/generated/numpy.ravel.html
from sklearn.model_selection import LeaveOneOut
from sklearn.decomposition import PCA
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

pca = PCA(.95)
X = pca.fit_transform(X_log)
print(X.shape)
Y = np.ravel(Y)

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
SVM = LinearSVC(multi_class = 'ovr', max_iter=10000)
cv = LeaveOneOut()

grid_search = GridSearchCV(SVM, param_grid=param_grid, cv=cv)
grid_search.fit(X, Y)

scores = cross_val_score(grid_search, X, Y, cv=cv)

print("Cross-validation scores: ", scores)
print("Mean cross-validation score: ", scores.mean())
print("Best estimator: {}".format(grid_search.best_estimator_))
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_predict.html#sklearn.model_selection.cross_val_predict
from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(grid_search, X, Y, cv=cv)
y_pred
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html
from sklearn.metrics import classification_report

print(classification_report(Y, y_pred, target_names=grid_search.classes_))
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html#sklearn.metrics.ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(Y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=grid_search.classes_)
disp.plot()
plt.show()
```
